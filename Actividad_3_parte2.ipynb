```python
#Verificar datos atipicos en la edad
# Crear un box plot
plt.boxplot(df['Edad'])
plt.title('Diagrama de Caja de Edad')
plt.ylabel('Edad')
plt.show()

# Calcular el IQR y los límites para detectar outliers
Q1 = df['Edad'].quantile(0.25)
Q3 = df['Edad'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers = df[(df['Edad'] < lower_bound) | (df['Edad'] > upper_bound)]
print("Valores atípicos en la variable Edad Grupo 1:", outliers)

# Crear un box plot
plt.boxplot(df2['Edad'])
plt.title('Diagrama de Caja de Edad')
plt.ylabel('Edad')
plt.show()

# Calcular el IQR y los límites para detectar outliers
Q1 = df2['Edad'].quantile(0.25)
Q3 = df2['Edad'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers = df2[(df2['Edad'] < lower_bound) | (df2['Edad'] > upper_bound)]
print("Valores atípicos en la variable Edad:", outliers)
```

```python
import scipy.stats as stats


# Coeficiente de correlación de Pearson
corr = df['Edad'].corr(df['Estatura'])
print("Coeficiente de correlación:", corr)

# Desviaciones estándar
std_edad = df['Edad'].std()
std_estatura = df['Estatura'].std()
print("Desviación estándar de la edad:", std_edad)
print("Desviación estándar de la estatura:", std_estatura)

# Test de Levene
stat, p = stats.levene(df['Edad'], df['Estatura'])
print('Test de Levene Statistic: %.4f' % stat)
print('p = %.4f' % p)
# Si p > 0.05, no se rechaza la hipótesis nula, y las varianzas son homogéneas

# Gráfico de dispersión
plt.scatter(df['Edad'], df['Estatura'])
plt.xlabel('Edad')
plt.ylabel('Estatura')
plt.title('Gráfico de Dispersión: Edad vs. Estatura grupo 1')
plt.show()

print('----------------------------------------------------------------------------')
# Coeficiente de correlación de Pearson
corr2 = df2['Edad'].corr(df2['Estatura'])
print("Coeficiente de correlación:", corr2)

# Desviaciones estándar
std_edad2 = df2['Edad'].std()
std_estatura2 = df2['Estatura'].std()
print("Desviación estándar de la edad:", std_edad2)
print("Desviación estándar de la estatura:", std_estatura2)

# Test de Levene
stat2, p = stats.levene(df2['Edad'], df2['Estatura'])
print('Test de Levene Statistic: %.4f' % stat2)
print('p = %.4f' % p)
# Si p > 0.05, no se rechaza la hipótesis nula, y las varianzas son homogéneas

# Gráfico de dispersión
plt.scatter(df2['Edad'], df2['Estatura'])
plt.xlabel('Edad')
plt.ylabel('Estatura')
plt.title('Gráfico de Dispersión: Edad vs. Estatura grupo 2')
plt.show()
```
