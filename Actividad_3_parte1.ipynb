```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

#Creamos nuestra base de datos para poder realizar el ejercicio.
# Definimos las posibles opciones para cada columna
sexos = ['M', 'F']
ciudades = ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Otras']

# Creamos una lista de diccionarios para almacenar los datos
data = []
for i in range(30): 
    # Generamos valores aleatorios para cada columna
    sexo = random.choice(sexos)
    edad = random.randint(18, 30)
    estatura = random.uniform(150, 190)
    nota = random.randint(0, 5)
    ciudad = random.choice(ciudades)

    # Agregamos un nuevo registro al conjunto de datos
    data.append({'Sexo': sexo, 'Edad': edad, 'Estatura': estatura, 'Nota': nota, 'Ciudad': ciudad})

# Creamos el DataFrame
df = pd.DataFrame(data)

# Mostramos las primeras 5 filas del DataFrame
print(df.head())
df = pd.DataFrame(data)
```

```python
# Tabla de frecuencias
freq_abs = df['Nota'].value_counts()
freq_rel = df['Nota'].value_counts(normalize=True)
print("Tabla de frecuencias absolutas:\n", freq_abs)
print("\nTabla de frecuencias relativas:\n", freq_rel)
```

```python
# Diagrama de barras para Ciudad
sns.countplot(x='Ciudad', data=df)
plt.title('Distribución de estudiantes por ciudad')
plt.show()
# Diagrama de sectores
plt.pie(df['Ciudad'].value_counts(), labels=df['Ciudad'].unique(), autopct='%1.1f%%')
plt.title('Proporción de estudiantes por ciudad')
plt.show()
```

```python
# Histograma
sns.histplot(data=df, x='Edad', bins=10, color='yellow')
plt.title('Distribución de edades')
plt.show()
```

```python

# Diagrama de caja y bigotes
sns.boxplot(x='Edad', data=df)
plt.title('Diagrama de caja y bigotes para Edad')
plt.show()
```

```python
# Identificando valores atípicos con range=1.5
Q1 = df['Edad'].quantile(0.25)
Q3 = df['Edad'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Edad'] < lower_bound) | (df['Edad'] > upper_bound)]
print("Valores atípicos:", outliers)

print('-----------------------------------')
# Identificando valores atípicos con range=0.5
Q1 = df['Edad'].quantile(0.25)
Q3 = df['Edad'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 0.5 * IQR
upper_bound = Q3 + 0.5 * IQR
outliers = df[(df['Edad'] < lower_bound) | (df['Edad'] > upper_bound)]
print("Valores atípicos:", outliers)

print('-----------------------------------')
```

```python
# Resumen general
resumen = df['Nota'].describe()
print(resumen)

# Calculando medidas individualmente
media = df['Nota'].mean()
mediana = df['Nota'].median()
desviacion_estandar = df['Nota'].std()

# Comparando resultados
print("Media calculada manualmente:", media)
print("Media según describe():", resumen['mean'])
```

```python
# Estatura media
estatura_media = df['Estatura'].mean()
print("Estatura media:", estatura_media)

# Medidas de dispersión
desviacion_estandar = df['Estatura'].std()
varianza = df['Estatura'].var()
print("Desviación estándar:", desviacion_estandar)
print("Varianza:", varianza)
```
